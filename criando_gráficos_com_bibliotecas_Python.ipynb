```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/canadian_immegration_data.csv")
df.set_index('Country', inplace=True)
anos = list(map(str,range(1980,2014)))
brasil = df.loc['Brazil',anos]
brasil_dict={'anos': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil=pd.DataFrame(brasil_dict)

plt.figure(figsize=(8,4))
plt.plot(dados_brasil['anos'], dados_brasil['imigrantes'])
plt.title('Imigração do Brasil para o Canada', fontsize= 18)
plt.xlabel('Anos')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show()
```

```
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['anos'], dados_brasil['imigrantes'], lw=3, marker='o', color='g')
ax.set_title('Imigração do Brasil para o Canada\n 1980 a 2013', fontsize=16, loc='left')
ax.set_xlabel('Anos')
ax.set_ylabel('Número de imigrantes')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.grid(linestyle='--')
plt.show()
```

```
fig, axs = plt.subplots(1, 2, figsize=(15,5))
axs[0].plot(dados_brasil['anos'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canada\n 1980 a 2013')
axs[0].set_xlabel('Anos')
axs[0].set_ylabel('Número de imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da Imigração do Brasil para o Canada\n 1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')
axs[1].grid()

plt.show()
```

```
fig, axs = plt.subplots(2, 2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)

axs[0,0].plot(df.loc['Brazil', anos])
axs[0,0].set_title('Brasil')
axs[0,1].plot(df.loc['Colombia', anos])
axs[0,1].set_title('Colômbia')
axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')
axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.grid()

for ax in axs.flat:
    ax.set(xlabel='Ano', ylabel='Número de imigrantes')

    ymin = 0
    ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)

plt.show()
```
