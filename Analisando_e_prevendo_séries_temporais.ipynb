```python
df_data.head()
```

```python
df_por_hora = df_data.groupby('horario')['contagem'].sum().reset_index()
plt.figure(figsize=(8,4))
sns.barplot(data=df_por_hora, x='horario', y='contagem', color='red')
plt.title('Bicicletas alugadas por hora')
```

```python
df_fim_de_semana = df_data[df_data['feriado']=='Sim']
alugueis_fim_de_semana = df_fim_de_semana.groupby('horario')['contagem'].sum().reset_index()
plt.figure(figsize=(8,4))
sns.barplot(data=alugueis_fim_de_semana, x='horario', y='contagem', color='green')
plt.title('Bicicletas alugadas no fim de semana')
```

```python
df_tempo = df_data.groupby(df_data['data_hora'].dt.date)['contagem'].sum().reset_index()
plt.figure(figsize=(12,4))
sns.lineplot(data=df_tempo, x='data_hora', y='contagem', color='blue')
plt.title('Bicicletas alugadas por dia')
plt.xlabel('Data')
plt.ylabel('Número de Aluguéis')
plt.show()
```

```python
alugueis_mensais = df_data.groupby('mes')['contagem'].sum().reset_index()
sns.barplot(data=alugueis_mensais, x='mes', y='contagem', color='orange')
plt.title('Bicicletas alugadas por mês')
plt.xlabel('Mês')
plt.ylabel('Número de Aluguéis')
plt.show()
```

```python
from prophet import Prophet
```

```python
df_prophet = df_data[['data_hora', 'contagem']].rename(columns={'data_hora':'ds', 'contagem':'y'})
df_prophet.head()
```

```python
df_propheft = df_prophet.groupby('ds')['y'].sum().reset_index()
```

```python
df_propheft
```

```python
df_prophet_daily = df_propheft.copy()
df_prophet_daily['ds'] = df_prophet_daily['ds'].dt.date
df_prophet_daily = df_prophet_daily.groupby('ds')['y'].sum().reset_index()
df_prophet_daily.head()
```

```python
import numpy as np
np.random.seed(3221)
modelo = Prophet()
modelo.fit(df_prophet)
futuro = modelo.make_future_dataframe(periods=90, freq='D')
previsoes = modelo.predict(futuro)
```

```python
fig1 = modelo.plot(previsoes)
```

```python
previsoes[['ds','yhat', 'yhat_lower','yhat_upper']]
```

```python
fig2 = modelo.plot_components(previsoes)
```

```python
df_treino = pd.DataFrame()
df_treino['ds'] = df_prophet['ds'][:3480]
df_treino['y'] = df_prophet['y'][:3480]
```

```python
df_teste = pd.DataFrame()
df_teste['ds'] = df_prophet['ds'][3480:]
df_teste['y'] = df_prophet['y'][3480:]
```

```python
np.random.seed(3221)
modelo = Prophet()
modelo.fit(df_treino)
futuro = modelo.make_future_dataframe(periods=150, freq='D')
previsoes = modelo.predict(futuro)
```

```python
fig1 =  modelo.plot(previsoes)
plt.plot(df_teste['ds'], df_teste['y'], 'o', color='red')
```

```python
df_previsoes = previsoes[['ds', 'yhat']]
df_comparacao = pd.merge(df_previsoes, df_teste, on='ds')
df_comparacao
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(df_comparacao['y'], df_comparacao['yhat'])
rmse = np.sqrt(mse)

print(f'MSE:{mse}, RMSE:{rmse}')
```
